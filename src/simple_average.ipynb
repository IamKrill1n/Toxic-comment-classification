```python
import pandas as pd
import numpy as np
import os
```

```python
# quick and dirty way to change the current working directory to root (/toxic-comment-classification)
# you should run this at least once just to be certain
from os import chdir, path, getcwd
if getcwd().endswith("src"):
    chdir(path.pardir)
if path.isfile("checkcwd"):
    print("Success")
else:
    raise Exception("Something went wrong. cwd=" + getcwd())
root_path = os.getcwd()
summision_path = root_path + '/kaggle/working'
```

```python
submission_files = os.listdir(summision_path)
label_cols = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
ensemble_submission = pd.DataFrame()

for file in submission_files:
    if file.endswith('.csv'):
        file_path = os.path.join(summision_path, file)
        submission_data = pd.read_csv(file_path)
        # print(file_path, submission_data.head())
        if ensemble_submission.empty:
            ensemble_submission = submission_data.copy()
            ensemble_submission[label_cols] = ensemble_submission[label_cols] * 0
            continue
        ensemble_submission[label_cols] = ensemble_submission[label_cols].add(submission_data[label_cols])
        ensemble_submission['id'] = submission_data['id']

print(len(submission_files))
ensemble_submission[label_cols] = ensemble_submission[label_cols] / len(submission_files)
ensemble_submission
# ensemble_submission.to_csv(summision_path + '/ensemble_submission.csv', index=False)
```

```python
ensemble_submission.to_csv(summision_path + '/average_submission.csv', index=False)
```
